__날짜 전처리__

```python
df['DATE'] = pd.to_datetime(df['DATE']).dt.date # datetime 형식으로 변환
```

__리뷰 전처리__

```python
df['REVIEW'] = df['TITLE'] + " " + df['REVIEW'] # 제목과 본문을 합쳐서 새로운 열 생성
del df['TITLE']
```

=========================================================================================================

```python
df = df.drop(columns=['Unnamed: 0', 'DISLIKE'], axis=1)
df = df[['DATE', 'STAR', 'REVIEW', 'LIKE']].sort_values(by=['DATE']) # 열 순서 조정, 날짜 기준으로 오름차순
```
