## 1.평점별로 빈도 분석

```python
import pickle
from pprint import pprint
import pandas as pd
```

```python
with open("D:/cleaned_data.pk", "rb") as f:
    data = pickle.load(f)
    
data.reset_index(drop=True, inplace=True)
print(data.head())
print(data.info())
```

```python
## 평점을 파악합니다.
star = set(data['STAR'])
star
```

```python
## 별점 별로 데이터를 묶습니다.
authors = data.groupby('STAR')
pprint(authors.groups)
print(type(authors.groups))
```

```python
# 별점 별로 묶은 데이터를 사용가능한 데이터 타입으로 바꿉니다.
author2doc = {}
for star , index in authors.groups.items():
    author2doc[star] = list(index)
author2doc
```

```python
# 별점 1에 대한 리뷰 내용 확인
star1_review = [str(data['REVIEW'][idx]) for idx in author2doc[1]]
star1_review
```

## 2.빈도분석

```python
# 전체 단어 분석을 위해 token들을 전부 모으기
import nltk

total_tokens = [token for doc in star1_review for token in doc.split()]
print(len(total_tokens))
```

```python
text = nltk.Text(total_tokens, name='NMSC')
print(len(set(text.tokens)))
pprint(text.vocab().most_common(10))
```

```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
%matplotlib inline

path = "c:/Windows/Fonts/malgun.ttf"
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')

plt.figure(figsize=(16, 10))
text.plot(50)
```
