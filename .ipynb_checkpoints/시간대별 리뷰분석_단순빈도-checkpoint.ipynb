## 1.평점별로 빈도 분석

```python
import pickle
from pprint import pprint
import pandas as pd
```

```python
with open("D:/cleaned_data.pk", "rb") as f:
    data = pickle.load(f)
    
data.reset_index(drop=True, inplace=True)
print(data.head())
print(data.info())
```

```python
## 4월을 기준으로 나누기
data_positive = data[data['STAR'] > 3] # 긍정 리뷰
data_negative = data[data['STAR'] < 3] # 부정 리뷰
```

```python
data_positive.head()
```

```python
## 나눈 시점부터 리뷰 가져오기
slice0 = [review for review in list(data_positive['REVIEW'])]
slice1 = [review for review in list(data_negative['REVIEW'])]
```

## 2.빈도분석

## 긍정리뷰

```python
# 전체 단어 분석을 위해 token들을 전부 모으기
import nltk

total_tokens = [token for doc in slice0 for token in doc.split()]
print(len(total_tokens))
```

```python
text = nltk.Text(total_tokens, name='NMSC')
print(len(set(text.tokens)))
pprint(text.vocab().most_common(10))
```
