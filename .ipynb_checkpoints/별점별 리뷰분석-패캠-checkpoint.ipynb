```python
import re
import MeCab # 형태소 분석기
mecab = MeCab.Tagger()
import requests
import pickle
import re
import ast
def mecab_nouns(text):
    nouns = []
    
    # 우리가 원하는 TOKEN\tPOS의 형태를 추출하는 정규표현식.
    pattern = re.compile(".*\t[A-Z]+")
    
    # 패턴에 맞는 문자열을 추출하여 konlpy의 mecab 결과와 같아지도록 수정.
    temp = [tuple(pattern.match(token).group(0).split("\t")) for token in mecab.parse(text).splitlines()[:-1]] 
    
    # 추출한 token중에 POS가 명사 분류에 속하는 토큰만 선택.
    for token in temp:
        if token[1] == "NNG" or token[1] == "NNP" or token[1] == "NNB" or token[1] == "NNBC" or token[1] == "NP" or token[1] == "NR":
            nouns.append(token[0])
    return nouns
```

```python


def message_cleaning(docs):

  
   
    # Series의 object를 str로 변경.
    docs = [str(doc) for doc in docs]
    
    
    # 2. 자음 모음 제거하기
    pattern2 = re.compile("[ㄱ-ㅎ]*[ㅏ-ㅢ]*")
    docs = [pattern2.sub("", doc) for doc in docs]
    # 3. 특수문자 제거
    pattern4 = re.compile("[\{\}\[\]\/?.,;:'|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]")
    docs = [pattern4.sub("", doc) for doc in docs]
    
    
    return docs

def define_stopwords(path):
    
    SW = set()
    # 불용어를 추가하는 방법 1.
    # SW.add("있다")
    
    # 불용어를 추가하는 방법 2.
    # stopwords-ko.txt에 직접 추가
    
    with open(path) as f:
        for word in f:
            SW.add(word)
            
    return SW

def text_tokenizer(doc):
#     return [word for word in mecab.morphs(doc) if word not in SW and len(word) > 1]
    
    # wordcloud를 위해 명사만 추출하는 경우.
    return [word for word in mecab_nouns(doc) if word not in SW and len(word) > 1]
```

```python
SW = define_stopwords("D:/연구알바/stopwords-ko.txt")
```

```python
from tqdm import tqdm_notebook # 시간 바
def text_tokenizing(corpus, tokenizer):
    # 명사 추출 / 형태소 분석 두 가지를 선택할 수 있게 만들어주는 함수를 만들어보자.
    mecab = MeCab.Tagger()
    token_corpus = []
    # tqdm을 사용하여 진행 과정을 볼 수 있게 만들어보자.
    if tokenizer == "noun":
        for n in tqdm_notebook(range(len(corpus))):
            token_text = mecab_nouns(corpus[n])
            token_text = [word for word in token_text if word not in SW and len(word) >1]
            token_corpus.append(token_text)
            
    elif tokenizer == "morph":
        for n in tqdm_notebook(range(len(corpus))):
            token_text = mecab.nouns(corpus[n])
            token_text = [word for word in token_text if word not in SW and len(word) >1]
            token_corpus.append(token_text)
    
    elif tokenizer == "word":
        for n in tqdm_notebook(range(len(corpus))):
            token_text = corpus[n].split()
            token_text = [word for word in token_text if word not in SW and len(word) >1]
            token_corpus.append(token_text)
            
    return token_corpus
```

```python
## 텍스트 정제
cleaned_corpus = message_cleaning(corpus)
print(cleaned_corpus)
```

```python
tokenized_text = text_tokenizing(cleaned_corpus,tokenizer = "noun")
print(tokenized_text)
```

```python
import pickle
from pprint import pprint
import pandas as pd
```

```python
with open("D:/cleaned_data.pk", "rb") as f:
    data = pickle.load(f)
    
data.reset_index(drop=True, inplace=True)
print(data.head())
print(data.info())
```

```python
## 평점을 파악합니다.
star = set(data['STAR'])
star
```

```python
## 별점 별로 데이터를 묶습니다.
authors = data.groupby('STAR')
pprint(authors.groups)
print(type(authors.groups))
```

```python
# 별점 별로 묶은 데이터를 사용가능한 데이터 타입으로 바꿉니다.
author2doc = {}
for star , index in authors.groups.items():
    author2doc[star] = list(index)
author2doc
```
