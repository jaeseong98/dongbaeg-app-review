```python
# 다양한 오차 측정 지표를 확인하기 위한 함수 정의

from sklearn.metrics import *

def get_clf_eval(y_test, pred=None, pred_proba=None): 
    confusion = confusion_matrix(y_test, pred) 
    accuracy = accuracy_score(y_test, pred) 
    precision = precision_score(y_test, pred) 
    recall = recall_score(y_test, pred) 
    f1 = f1_score(y_test, pred) 
    # ROC AUC 
    roc_auc = roc_auc_score(y_test, pred_proba) 
    
    print('Confusion Matrix') 
    print(confusion) 
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, f1스코어: {3:.4f}, roc_auc: {4:.4f}'.format( accuracy, precision, recall, f1, roc_auc))
```

 # 전처리


### 1. 이상치제거 - Duration과 Amount

```python
#이상치 값 제거하는 함수정의(책 참고)
def get_outlier(df=None, column=None, weight=1.5):
    
  # target 값과 상관관계가 높은 열을 우선적으로 진행
    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)

    IQR = quantile_75 - quantile_25
    IQR_weight = IQR*weight
  
    lowest = quantile_25 - IQR_weight
    highest = quantile_75 + IQR_weight
  
    outlier_idx = df[column][ (df[column] < lowest) | (df[column] > highest) ].index
    return outlier_idx

# 함수 사용해서 이상치 값 삭제
outlier_idx_duration = get_outlier(df=df, column='DURATION', weight=1.5)
df.drop(outlier_idx_duration, axis=0, inplace=True)
outlier_idx_amount = get_outlier(df=df, column='AMOUNT', weight=1.5)
df.drop(outlier_idx_amount, axis=0, inplace=True)
```

## 2. t-test로 삭제할 컬럼 찾기

```python
import scipy.stats as stats
from scipy.stats import f
from scipy.stats import ttest_ind
table = pd.DataFrame(data = {'Variable':[], 'Delete':[],'Equal/Not Equal':[], 'p-value for f-test':[], 'p-value for t-test':[]})
for i in range(1,df.shape[1]):
    lresult = stats.levene(df.iloc[:,0], df.iloc[:,i])
    f_value = pd.DataFrame(lresult)
    if f_value.iloc[1,0] > 0.05:
        abc = pd.DataFrame(ttest_ind(df.iloc[:,0], df.iloc[:,i], equal_var = True))
        if abc.iloc[1,0] > 0.05:
            table.loc[i] = [df.iloc[:,i].name, 'delete', 'equal', f_value.iloc[1,0], abc.iloc[1,0]]
        else:
            table.loc[i] = [df.iloc[:,i].name, 'not delete', 'equal', f_value.iloc[1,0], abc.iloc[1,0]]
    else:
        cba = pd.DataFrame(ttest_ind(df.iloc[:,0], df.iloc[:,i], equal_var = False))
        if cba.iloc[1,0] > 0.05:
            table.loc[i] = [df.iloc[:,i].name, 'delete', 'not equal', f_value.iloc[1,0], cba.iloc[1,0]]
        else:
            table.loc[i] = [df.iloc[:,i].name, 'not delete', 'not equal', f_value.iloc[1,0], cba.iloc[1,0]]
            
table
```

→ table의 Delete 열에서 "delete"이라고 나온 변수 "OWN_RES" drop
